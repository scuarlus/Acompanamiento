```python
import pandas as pd

def descargar_y_analizar_dataset():
    # URL del dataset CSV de "Tiendas naturistas, farmacias y droguerías"
    url_csv = 'https://www.datos.gov.co/resource/8sgh-9kiy.csv'

    print("Descargando dataset desde:", url_csv)
    try:
        df = pd.read_csv(url_csv)
        print("Carga exitosa.\n")
    except Exception as e:
        print("Error al descargar o cargar el dataset:", e)
        return

    print("=== Primeras 5 filas ===")
    print(df.head(), "\n")

    print("=== Información del DataFrame ===")
    print(df.info(), "\n")

    print("=== Conteo de valores nulos por columna ===")
    print(df.isnull().sum(), "\n")

    print("=== Estadísticas descriptivas (solo numéricas) ===")
    print(df.describe(), "\n")

    print("=== Columnas disponibles ===")
    print(df.columns.tolist(), "\n")

    readme_text = """
    README - Dataset: Tiendas Naturistas / Farmacias / Droguerías

    Descripción:
    Listado de establecimientos clasificados como tiendas naturistas, farmacias o droguerías en Colombia.
    Incluye datos sobre nombre, ubicación, tipo de establecimiento y más.

    Fuente:
    Portal Datos Abiertos Colombia - Salud y Protección Social
    URL: https://www.datos.gov.co/resource/8sgh-9kiy.csv

    Fecha de última actualización:
    20 de abril de 2024

    Licencia:
    Consultar en la página oficial (posiblemente CC BY 4.0).

    Uso:
    Este dataset es útil para análisis de distribución geográfica y comercialización de productos farmacéuticos y naturistas.

    Consideraciones:
    - Puede contener valores nulos.
    - Revisar tipos de datos para limpieza o análisis específicos.
    """

    print(readme_text)

if __name__ == "__main__":
    descargar_y_analizar_dataset()
```
